```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
from sklearn.naive_bayes import GaussianNB
model_params = {
#     'svm': {
#         'model': SVC(),
#         'params' : {
#             'C': [1,10,20],
#             'kernel': ['rbf','linear']
#         }  
#     },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params' : {
            'n_estimators': [1,2,3,4,5],
            'max_depth':[25,30,35]
        }
    },
    'D_tree':{
        'model':tree.DecisionTreeClassifier(),
        'params':{
            'criterion':['gini','entropy']
        }
    },
#     'Gaussian_nb':{
#         'model':GaussianNB(),
#         'params':{}
#     },
}
```

```python
scores = []

for model_name, mp in model_params.items():
    clf =  GridSearchCV(mp['model'], mp['params'], return_train_score=False)
    clf.fit(X_train,y_train)
    scores.append({
        'model': model_name,
#         'best_score': clf.best_score_,
#         'best_params': clf.best_params_
#         'mean_test_score':clf.mean_test_score,
        'mean_fit_time':clf.mean_fit_time_, 
        'mean_score_time':clf.mean_score_time_,
        'params':clf.params_
    })
    
df = pd.DataFrame(scores,columns=['model','mean_test_score','mean_fit_time','mean_score_time','params'])
df
```
